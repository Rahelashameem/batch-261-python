# Call by Value

-  In call by value, a parameter acts within the function as a new local variable initialized to the value of the argument (a      local (isolated) copy of the argument).

-  When Immutable objects such as whole numbers, strings, etc are passed as arguments to the function call, the it can be          considered as Call by Value.

-  when the values are modified within the function, then the changes do not get reflected outside the function.


## Syntax

-   function_name(variable_name1,variable_name2...)

-   The arguments are copied to the function parameter.

-   Changes made inside the functions are not reflected.

-    A copy of a variable is passed.

-   Immutable Objects are passed in arguments.

-   While calling the function we pass values and called as call by values.

-   The value of each variable is calling the function and copy variable into the variables of the function.

-   The changes are made to dummy variables in the called function that have no effect on the actual variable of the function.

-   We cannot change the values of the actual variable through a function call.

-   change in a copy of a variable doesnot modify the value of original variable

-   The values are passed by the simple techniques


```python
### Python code to demonstrate 
### call by value
 
 
string = "Hello World"
 
 
def test(string):
     
    string = "Hello World hi"
    print("Inside Function:", string)
     
        
test(string)
print("Outside Function:", string)
```

# Call by reference

-  It is a way of passing arguments to a function call in which both the actual argument and formal parameters refer to the same    memory locations

-  And any changes made within the function are reflected in the actual parameters of the function.

-  When Mutable objects such as list, dict, set, etc., are passed as arguments to the function call, it can be called Call by      reference 

-  When the values are modified within the function, the change also gets reflected outside the function.

    -   Actual parameters - The parameters passed to a function
    - Formal parameters - The parameters received by a function

## Syntax

-   function_name(&variable_name1, &variable_name2...)

```python
# Python code to demonstrate 
# call by reference
 
 
def add_more(list):
    list.append(50)
    print("Inside Function", list)
 
# Driver's code
mylist = [10,20,30,40]
 
add_more(mylist)
print("Outside Function:", mylist)
```

```python
-   The arguments, as well as the formal parameters, refer to the same location.

-   Changes made inside the functions are reflected.

-   Mutable objects are passed in arguments

-   A variable itself is passed

-   Change in a copy of a variable modify the value of original variable.

-   At the time of calling instead of passing the values we pass the address of the variables that means the location of variables 

-   The address of variable in function while calling is copied into dummy variables of a function.

-   By using the address we have the access to the actual variable.

-   We can change the values of the actual variable through a function call.

-   The pointers are necessary to define and store the address of variables.
```

### Difference between call by value & call by reference:

-   In call by value, a parameter acts within the function as a new local variable initialized to the value of the argument (a       local (isolated) copy of the argument).

-   In call by reference, the argument variable supplied by the caller can be affected by actions within the called function.
