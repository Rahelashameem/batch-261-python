## slicing 

-    To access a range of items in a list, you need to slice a list. 
-    One way to do this is to use the simple slicing operator :
-    With this operator you can specify where to start the slicing, where to end and specify the step.
-    If L is a list, the expression L [ start , stop , step ] returns the portion of the list from index start to index stop, at      a step size step.
-    slicing in Python is a common practice and can be used both with positive indexes as well as negative indexes
### syntax:

  

- #### start ---> Defines as the index of the list where the slicing needs to start.
- #### stop ----> Defines as the index of the list where the slicing wants to ends. 
- #### step ----> Defines as the nth item that you want to allow and select within the specified range start to stop.

### positive slicing (left to right)

-  if step size is positive then,
-  start value is 0
-  end value is length of the datatype
-  stepsize is +1


```python
#### Example of positive slicing

L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[2:7])
```

```python
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'] ## Note that the item at index 7 'h' is not included
print(L[2:7])
```

```python
### Negative slicing (right to left)

-  if step size is negative then,
-  start value is -1
-  end value is -(n+1)
-  stepsize is -1
```

```python
#### Example of Negative slicing

L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[-7:-2])
```

```python
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[-7:-2])
```

## copy()

-     Copy() methods return a shallow copy of a list, which returns a new list without modifying the original lists.

-     Sometimes, there is a need to reuse any object, hence copy methods are always of great utility. 

### Syntax: 

-     list.copy()

### Parameters: 

-     The copy() method doesnâ€™t take any parameters.

### Returns: 

-    Returns a shallow copy of a list. A shallow copy means any modification in the new list wonâ€™t be reflected in the              original list.

```python
help(z.copy)
```

```python
x = [11, 22, 33]
print(x, hex(id(x)))

y = x
print(y, hex(id(y)))

print(hex(id(x[0])), hex(id(y[0])))
```

```python
z = x.copy()
print(z, hex(id(z)))
```

```python
x = [11, 22, 33]
print(x, hex(id(x)))

y = x
print(y, hex(id(y))) # x &y have same memory location

x.append(44)## append 44
print(y)
print(hex(id(x[0])), hex(id(y[0]))) #same memory location
```

```python
z = x.copy()
x.append(55) ## append 55
print(z, hex(id(z)))
```

```python
a = [1, [4,5,6]]
print(a, a[1], hex(id(a)), hex(id(a[1])))

b = a.copy() #same memory location 
print(b, b[1], hex(id(b)), hex(id(b[1])))
```

## Deep copy()

-    A deep copy creates a new compound object before inserting copies of the items found in the original into it in a              recursive  manner.

-    In the case of deep copy, a copy of the object is copied into another object.

-    It means that any changes made to a copy of the object do not reflect in the original object.

-    It is used to copy nested list

```python
import copy
from copy import deepcopy # to handle nested list deep copy is used

a = [1, [4,5,6]]
b = deepcopy(a)
print(a, id(a), id(a[1])) # different memory location
print(b, id(b), id(b[1]))
```

```python
a[1].append(7)
a
```
