# Tuples

-    Tuple is an immutable ordered sequence of items.

-    Tuple is a collection of objects separated by commas.

-    Tuples are used to store multiple items in a single variable.

-    Tuples are written with round brackets.

-    A tuple can have any number of items and they may be of different types (integer, float, list, string, etc.).

-    In some ways, a tuple is similar to a Python list in terms of indexing, nested objects, and repetition but the main 

     difference between both is Python tuple is immutable, unlike the Python list which is mutable. 

## Tuple Items

-      Tuple items are ordered, unchangeable, and allow duplicate values.

-      Tuple items are indexed, the first item has index [0], the second item has index [1] etc.

## Ordered

-     When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.



## Unchangeable

-       Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.



## Allow Duplicates

-        Since tuples are indexed, they can have items with the same value:

## Tuples are Immutable

-     Tuples in Python are similar to Python lists but not entirely,tuples are immutable and ordered and allow duplicate             values. 

-     Some Characteristics of Tuples in Python.

-     We can find items in a tuple since finding any item does not make changes in the tuple.

-     One cannot add items to a tuple once it is created. 

-     Tuples cannot be appended or extended.

-     We cannot remove items from a tuple once it is created.

```python

r = (1, 2, 3, 4, 5)
 
# tuples are indexed
print(r[1])
print(r[4])
 
# tuples contain duplicate elements
r = (1, 2, 3, 4, 2, 3)
print(r)
 
# adding an element
r[1] = 100
print(r)
```

## Accessing Values in  Tuples

-   To access values or elements in tuple, we can use indexing or slicing.

-   Indexing starts from 0, which means the first element has an index of 0, the second element has an index of 1, and so on.

-   We can use negative indexing, which starts from the last element (-1), the second last element (-2), and so on.

## Indexing in tuple

-       Indexing is a technique for accessing the elements of a tuple. 

-       There are various ways by which we can access an element of a tuple.


### Positive Index

-      In the case of Positive Indexing, the first element of the list has the index number 0, 

       and the last element of the list has the index number N-1, 
       
       where N is the total number of elements in the tuple (size of the tuple).

```python
n = ("Geeks", "for", "Geeks")
 
print(n[0])
print(n[1])
print(n[2])
```

## Negative Indexing

-  Index -1 represents the last element and -N represents the first element of the tuple, where N is the length of the tuple

```python
c = (1, 2, 3)
 
print(c[-1])
print(c[-2])
print(c[-3])
```

## Slicing

-   Syntax tuple[ Initial : End : IndexJump ]

-   If t is a tuple, then the above expression returns the portion of the tuple from index Initial to index End, at a step size     IndexJump

-   It can be both with positive indexes as well as negative indexes.

```python
s= (5,70,37,20,88,10,50)
print(s,type(s),id(s),sep='\n')
print(s[1:4])
```

## Concatenation of Tuples

-   Concatenation of tuple is the process of joining two or more Tuples.

-   Concatenation is done by the use of ‘+’ operator.

-   Concatenation of tuples is done always from the end of the original tuple.

-   Other arithmetic operations do not apply on Tuples.

-   Only the same datatypes can be combined with concatenation, an error arises if a list and a tuple are combined.

## Unpacking a Tuple

-      When we create a tuple, we normally assign values to it. This is called "packing" a tuple.

-      But, in Python, we are also allowed to extract the values back into variables. This is called "unpacking"

```python
### Packing a tuple:

fruits = ("apple", "banana", "cherry")

### Unpacking a tuple:

fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

## Methods in Tuple

-     1. index( )

         Find in the tuple and returns the index of the given value where it’s available
         
-      2. count( )

          Returns the frequency of occurrence of a specified value

## 1. Count()

-       The count() method of Tuple returns the number of times the given element appears in the tuple.

### Syntax:

-       tuple.count(element)

        here the element is the element that is to be counted
        
### Parameters:

-       element: is the item whose count is to be returned.

-       Returns: Returns the count of how many times element occurs in the tuple.

```python
t=(1,2,2,4,2)
t.count(2)
```

## Index()

-     The Index() method returns the first occurrence of the given element from the tuple.

### Syntax:

-     tuple.index(element, start, end)

### Parameters:

-    element: The element to be searched.

-    start (Optional): The starting index from where the searching is started

-    end (Optional): The ending index till where the searching is done

-    Note: This method raises a ValueError if the element is not found in the tuple.

```python
t=(0,7,4,2,6,9)
t.index(2)
```
